```python
public_key_df1       ='https://disk.yandex.ru/d/c97ZgkdB0qPoqQ'
public_key_df2       ='https://disk.yandex.ru/d/JnAaaUjZeXqbDA'
public_key_df_report1='https://disk.yandex.ru/d/xpGQGQX6piZ6Kg'
public_key_df_report2='https://disk.yandex.ru/d/zyUoqOIqmY7D7w'
```

```python
final_url_for_df1    =base_url + urlencode(dict(public_key=public_key_df1))
final_url_for_df2    =base_url + urlencode(dict(public_key=public_key_df2))
final_url_for_report1=base_url + urlencode(dict(public_key=public_key_df_report1))
final_url_for_report2=base_url + urlencode(dict(public_key=public_key_df_report2))
```

```python
df1_response       =requests.get(final_url_for_df1)
df2_response       =requests.get(final_url_for_df2)
df_report1_response=requests.get(final_url_for_report1)
df_report2_response=requests.get(final_url_for_report2)

print(df1_response)
print(df2_response)
print(df_report1_response)
print(df_report2_response)
#response<300, значит, все сработало корректно и без ошибок, ссылки создались
```

```python
download_url_for_df1         = df1_response.json()['href']
download_url_for_df2         = df2_response.json()['href']
download_url_for_df_report1  = df_report1_response.json()['href']
download_url_for_df_report2  = df_report2_response.json()['href']
```

```python
# загружаем файлы в df
df1       =pd.read_csv(download_url_for_df1,sep=',')
df2       =pd.read_csv(download_url_for_df2,sep=',')
df_report1=pd.read_csv(download_url_for_df_report1, sep=',')
df_report2=pd.read_csv(download_url_for_df_report2, sep=',')
```

# **1. Объедини каждый вид файлов, чтобы получить все данные**

Объединим датафреймы скачивания приложений

```python
df_downloads=pd.concat([df1,df2],ignore_index=True)
```

Исходя из данных датафрейма со скачиванием приложения, нет полей где было бы указано бинарное значение(фрод/ не фрод), поэтому будем считать показатель фрода, используя агрегированные данные отчета.

```python
downloads_count=df_downloads.groupby('Media Source', as_index=False).agg({'Event Name':'count'}).sort_values(by="Event Name", ascending=False)
downloads_count.head()
```

Объединим данные датафремов отчетов о фродах

```python
frod_report_final=pd.concat([df_rep2,df_rep2],ignore_index=True)
frod_report_final
```

Посмотрим на уникальные значения столбца Attributions

```python
frod_report_final['Attributions'].unique()
```

```python
frod_report_final.query('Attributions == "10\xa0302"')
```

Такое иногда бывает, что пробел странно считывается, поэтому заменим в строках часть текста между цифр '\xa0' на пустое значение, а затем конвертируем из текста в числа с плавающей точкой(float64)

```python
frod_report_final['Attributions'] = frod_report_final['Attributions'].str.replace('\xa0', '')
```

Проверим результат преобразований столбца

```python
frod_report_final['Attributions'].unique()
```

Удалим дубликаты в датафрейме

```python
frod_report_final=frod_report_final.drop_duplicates()
```

Поскольку процент фрода по источникам считается из пропорции,необходимо использовать простой джойн, включающий в себя пересечения данных в двух датафреймах. В обоих датафреймах есть источники, данных о которых пропущенны

```python
merged_df=downloads_count.merge(frod_report_final, left_on="Media Source", right_on="Media source", how='inner')
```

```python
merged_df=merged_df[['Media source', 'Event Name','Attributions','Total fraudulent attribution']]
```

Заменим значения столбца, где есть '\xa0' на пустое значение и конвертируем в числовой

```python
merged_df['Total fraudulent attribution1']=merged_df['Total fraudulent attribution'].str.replace('\xa0', '')
```

```python
merged_df['Event Name']=merged_df['Event Name'].astype(float)
merged_df['Attributions']=merged_df['Attributions'].astype(float)
merged_df['Total fraudulent attribution1']=merged_df['Total fraudulent attribution1'].astype(float)
```

```python
merged_df=merged_df[['Media source', 'Event Name', 'Attributions', 'Total fraudulent attribution1']]
```

```python
merged_df.dtypes
```

Типы данных соответствуют, можем производить необходимые расчеты

```python
merged_df.head(2)
```

# **2. По источникам (Media source) рассчитай  кол-во скачиваний всего и кол-во фродовых скачиваний, выведи процентное содержание форда по источникам**

Посчитаем количество фрода и его процент по источникам

```python
merged_df['fraud_quality']=((merged_df['Event Name']*merged_df['Total fraudulent attribution1'])/merged_df['Attributions']).round(0)
merged_df['fraud_percent_count']=((merged_df['fraud_quality']/merged_df['Event Name'])*100).round(2)
```

**Ответ на задание 2**

```python
merged_df[['Media source','Event Name','fraud_quality','fraud_percent_count']]
```

# **3.1По каждому обычному инсталлу присвой цену 0,5 доллара (наша прибыль), каждому фродовому -0,2 доллара (рекламодатель вычитает у нас)**

**Ответ на задание 3.1**

```python
merged_df['profit']=(merged_df['Event Name']*0.5- merged_df['fraud_quality']*0.2).round(2)
merged_df['revenue']=(merged_df['Event Name']*0.5).round(2)
merged_df['fraud_losses']=(merged_df['fraud_quality']*0.2).round(2)
```

# **3.2 Выведи топ-5 источников по фродовым скачиваниям (отдельно с самым высоким и низким процентом)**

Ответ на задание 3.2

**Топ-5 по фродовым скачиваниям(с самым высоким процентом)**

```python
merged_df[['Media source','fraud_quality','profit', 'fraud_losses',	'fraud_percent_count','revenue']].sort_values(by='fraud_percent_count', ascending=False).head(5)
```

**Топ-5 по фродовым скачиваниям( с самым низким процентом)**

```python
merged_df[['Media source','fraud_quality','profit', 'fraud_losses',	'fraud_percent_count','revenue']].sort_values(by='fraud_percent_count', ascending=False).tail(5)
```

# **4. Выведи суммарно сколько денег мы заработали, сколько заплатили за фрод и разницу между ними**

**Ответ на задание 4**

```python
merged_df[['profit','fraud_losses', 'revenue']].sum()
```

# **5.Выведи Топ-5 источников с самым высоким содержанием фрода, сколько суммарно мы заплатили за этот фрод и какой процент от всего фрода составляют эти 5 источников**

Запишем в new_df данные по источникам с самым высоким процентом фрода

```python
new_df=merged_df[['Media source','fraud_quality','profit', 'fraud_losses',	'fraud_percent_count','revenue']].sort_values(by='fraud_percent_count', ascending=False).head(5)
```

Посчитаем процент фрода у каждого источника относительно всех

```python
new_df['percent_of_fraud']=((new_df['fraud_percent_count']/ merged_df['fraud_percent_count'].sum())*100).round(4)
```

Поскольку у топа-5 источников одинаковый процент фрода и составляет 100%, поле percent_of_fraud будет иметь одинаковое значение

```python
new_df['fraud_percent_count']
```

Посчитаем сколько всего убытков на пять источников с самым высоким содержанием фрода и какой процент фрода от всех источников суммарно на 5 шт

```python
new_df['total_fraud_losses']=new_df['fraud_losses'].sum()
new_df['total_per_fraud']=new_df['percent_of_fraud'].sum()
```

 **Ответ на задание 5**

```python
new_df[['Media source','fraud_losses','percent_of_fraud','total_fraud_losses','total_per_fraud']]
```
